```python
import plotly
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../brew/brew_data-cask.tsv", sep="\t")
```

```python
list(df)
```

```python
col_name = "analytics|install|30d|command"
fig = px.histogram(df, x=col_name, log_y=True, title="Installs in 30 days")
fig.show()
df.sort_values([col_name], ascending=[False]).head(5)[["full_token", col_name]]
```
