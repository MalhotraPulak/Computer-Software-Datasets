```python
import requests
```

```python
data = requests.get("https://formulae.brew.sh/api/formula.json")
```

```python
names = [d["name"] for d in data.json()]
```

```python
names
```

```python
big_json = {}
for idx, name in enumerate(names):
    print("Getting " + name + " " + str(idx))
    request_url = f"https://formulae.brew.sh/api/formula/{name}.json"
    d = requests.get(request_url)
    if (d.status_code == 404):
        print("not found for " + name)
        continue
    big_json[name] = d.json()
```

```python
import json
with open("data.json", "w+") as f:
    json.dump(big_json, f)
```

```python
with open("data_indented.json", "w+") as f:
    json.dump(big_json, f, indent=2)
```
