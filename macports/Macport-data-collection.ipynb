```python
import subprocess
with open("ports_list.txt", "r") as r:
  ports = [f.split()[0] for f in r.readlines()]
ports
```

```python
from tqdm import tqdm
import time
import json
def process(ports, num):
  def toJson(details):
    ans = {}
    lines = details.split("\n")
    ans["port-name"], ans["ver"], ans["category"] = lines[0].split(" ", 2) 
    for line in lines[1:]:
      if ":" in line:
        key = line.split(":", 1)[0].strip()
        value = line.split(":", 1)[1].strip()
        ans[key] = value
    return ans

  data = []
  for idx in tqdm(range(len(ports))):
      port = ports[idx]
      command = f"port info {port}"
      process = subprocess.Popen(command.split(), stdout=subprocess.PIPE)
      output, error = process.communicate()
      details = output.decode("utf-8")
      details = toJson(details)
      data.append(details)

  with open(f"data/port-{num}", "w") as f:
    json.dump(data, f)
```

```python
delta = 1000
for batch_no, start_idx in enumerate(range(0, len(ports), delta)):
  print(batch_no, start_idx, delta)
  process(ports[start_idx: start_idx+delta], batch_no)
```
